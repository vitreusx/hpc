```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("stencil.out", header=None)
df.columns = ["rep", "size", "R", "algo", "time"]
df
```

```python
exp_groups = df.groupby(["size", "R", "algo"])
exp_df = exp_groups.agg({"time": ["mean", "std"]}).reset_index()
exp_df
```

```python
fin_df = exp_df[(exp_df["size"] == int(64e6)) & (exp_df["R"] == int(3e3))]
fin_df.sort_values(by=("time", "mean"))
```
