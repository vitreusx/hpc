```python
import pandas as pd
import pandasql as ps
import plotly.graph_objects as go
import numpy as np
```

```python
df_v4 = pd.read_csv("results-v4.csv")
df_v4
```

```python
ps.sqldf("""
select size, num_streams, mean
from (
    select size, num_streams, block_size, avg(dur) mean
    from df_v4
    where section = 'no-alloc'
    group by size, num_streams, block_size)
group by size, num_streams
having mean = min(mean);
""")
```
