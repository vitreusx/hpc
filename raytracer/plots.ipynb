```python
import pandas as pd
import pandasql as ps
import plotly.graph_objects as go
import numpy as np
```

```python
df = pd.read_csv("results.csv")
df
```

```python
total_df = ps.sqldf("""
select algo, x, y, avg(dur) mean, sqrt(avg(dur*dur)-avg(dur)*avg(dur)) std
from df
where section is null or section == 'total'
group by algo, x, y;
""")
total_df
```

```python
hypers_df = ps.sqldf("""
select algo, mean, std
from total_df
group by algo
having mean = min(mean)
order by algo;
""")
hypers_df
```
